## 4. Analysis after addition of new features

```python
# Code Reference: https://www.geeksforgeeks.org/countplot-using-seaborn-in-python/

def bar_plot(df, feat, head, label1, fig_1):
  '''Plot bar graphs for a feature passed as argument'''

  df1 = df.groupby([feat])[feat].count()                                    # Group the feature and then Count them
  df1 = df1.reset_index(name='Count')                                       # Reset Index to get two columns as Feature name and Count

  # Code Reference: https://www.geeksforgeeks.org/display-percentage-above-bar-chart-in-matplotlib/
  percentage = []
  total = df1['Count'].sum()
  for i in range(df1.shape[0]):
    pct = (df1.Count[i] / total) * 100                                      # compute percentage of each format
    percentage.append(round(pct,2))
  df1['Percentage'] = percentage                                            # store the computed percentage into dataframe

  plt.figure(figsize= fig_1)
  sns.set_style("ticks",{'axes.grid' : True})
  graph = sns.barplot(x = df1[feat], y = df1['Count'])                      # Plot a Bar graph for a features with respect to its count
  plt.title('Count Plot for '+ head)                                        # Add title of the plot
  plt.xlabel(label1)                                                        # Add label of x-axis
  plt.ylabel('Count')                                                       # Add label of y-axis

  i = 0
  for p in graph.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()                                                       # Show percentage as text on top and centre of bar graph
    plt.text(x+width/2, y+height*1.01, str(df1.Percentage[i])+'%', ha='center', weight='bold')      
    i += 1
  plt.show()
```

### 4.3. Add New Feature - Days of Filed Claims

```python
# Code Reference: https://www.geeksforgeeks.org/get-month-and-year-from-date-in-pandas-python/

# Add new feature number of days of applied claim of the admitted beneficiary
train_inp['Claim_Days'] = (pd.DatetimeIndex(train_inp['ClaimEndDt']) - pd.DatetimeIndex(train_inp['ClaimStartDt'])).days
train_inp.head()
```

```python
df1 = train_inp.groupby('Claim_Days')['Claim_Days'].count()                              # Group the Claim_Days feature and then Count them
df1 = df1.reset_index(name='Count')                                                      # Reset Index to get two columns as Claim_Days and Count
total = df1['Count'].sum()
if len(df1)>=30:
  df1 = df1.sort_values('Count', ascending=False)
  df1 = df1[0:30]                                                                        # Store Top 30 values of the  Claim_Days feature by count
  df1 = df1.reset_index(drop= True)                                                      # Reset Index to get two columns as Claim_Days and Count

print(df1.head(), '\n')
percentage = []
for i in range(df1.shape[0]):
  pct = (df1.Count[i] / total) * 100                                                     # compute percentage of top 30 values from total Count
  percentage.append(round(pct,2))
df1['Percentage'] = percentage                                                           # store the computed percentage into dataframe
  
plt.figure(figsize= (20, 8))
sns.set_style("ticks",{'axes.grid' : True})                                              # Plot a Bar graph for Top-30 Claim_Days with respect to its count
graph = sns.barplot(x = df1['Claim_Days'], y = df1['Count'], order=df1.sort_values('Count', ascending=False).Claim_Days) 
plt.title('Count Plot for Days of Claims Applied(Top-30)')                               # Add title of the plot
plt.xlabel('Number of Days of Applied Claims')                                           # Add label of x-axis
plt.ylabel('Count')                                                                      # Add label of y-axis

i = 0
for p in graph.patches:
  width = p.get_width()
  height = p.get_height()
  x, y = p.get_xy()                                                                      # Show percentage as text on top and centre of bar graph
  plt.text(x+width/2, y+height*1.01, str(df1.Percentage[i])+'%', ha='center', weight='bold')  
  i += 1
plt.show()
```

#### <b><u>Observations</b></u>
1. <b>17.05%</b> of the claims have been applied for <b>3 days</b>.
2. Only <b>2 claims</b> have been applied for <b>36 days</b>.
3. <b>Why Claim Days?</b>
<ol type = 'a'>
<li>To check maximum claims have been filed for how many days?</li> 
<li>To get more idea about claim days, we need to check the hospitalization days and claims days. They should be nearly similar to avoid any frauds.</li>
<li>For example: If <b>Claim Days> Hospitalization Days</b> then it is a <b>fraud</b>.</li>
<li>So, it can help us detect, how many claims have Claim Days > Hospitalization Days?</li>
</ol>

<b><u>Point of Interest</u>:</b> Claim Days should be considered as a feature to check its impact after merging the data with class labels. It helps us in identifying the false claims, which might occur due to forging in Claim Start and End Dates.

#### <b>Q: What is the reimbursement amount granted when the claims filed were less than 3/deductible amount paid was NULL?</b> 

```python
total_reimb_amt = train_out['InscClaimAmtReimbursed'].sum()                       # Sum of total reimbursement amount spend on the applied claims
print('Total Reimbursement Amount = 148 Million')
print('Total Claims Applied = 517K')
# Less than 3 claims account for what percent of the total claims?
avg_lessthan3_claim = round((prv_claims_reimb_amt[prv_claims_reimb_amt['Total_Claims']<3]['Total_Claims'].sum()/prv_claims_reimb_amt['Total_Claims'].sum())*100,2)
# Average reimbursement amount spend on the less than 3 claims
avg_lessthan3_claim_amt = round((prv_claims_reimb_amt[prv_claims_reimb_amt['Total_Claims'] < 3] ['Total_Reimbursed_Amount'].sum()/total_reimb_amt)*100,2)
print('{}% of Total Reimbursement Amount(148 Million) has been spent on {}% of the Total Claims(517K)'.format(avg_lessthan3_claim_amt, avg_lessthan3_claim))

# Annual Deductions will NULL values
avg_deduct_claim = round((train_out[train_out['DeductibleAmtPaid'].isna()]['ClaimID'].count()/train_out['DeductibleAmtPaid'].count())*100,2)
# What percent of the total reimbursement amount is paid to those with NULL deductions annually?
avg_reimb_deduct = round((train_out[train_out['DeductibleAmtPaid'].isna()]['InscClaimAmtReimbursed'].sum()/train_out['InscClaimAmtReimbursed'].sum())*100,2)
print('{}% of Total Payments have NULL Annual Deductions but {}% Total Reimbursements take place in such cases.'.format(avg_deduct_claim, avg_reimb_deduct))
```

3. <b>No Deductions are NULL</b> which are not paid in the form of premiums. This implies that <b>No Money</b> has been granted to those <u>beneficiaries who have not paid any money as premiums</u> respectively.
