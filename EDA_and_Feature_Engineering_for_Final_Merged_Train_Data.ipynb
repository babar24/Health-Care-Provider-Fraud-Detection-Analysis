## 3. Train Data

### 3.1 Load Train Data

```python
train_ben = pd.read_csv('Train_Beneficiarydata-1542865627584.csv')          # Load the Train Beneficiary Data
train_inp = pd.read_csv('Train_Inpatientdata-1542865627584.csv')            # Load the Train InPatient Data
train_out = pd.read_csv('Train_Outpatientdata-1542865627584.csv')           # Load the Train OutPatient Data
train_labels = pd.read_csv('Train-1542865627584.csv')                       # Load the Train Class Labels Data
```

### 3.2 Add New Feature - Is Hospitalised?

<b>Before Merging all the Train Data</b>

*   For InPatients --> Value = 1(Hospitalised)
*   For OutPatients --> Value = 0(Not Hospitalised)  



```python
# Code Reference: https://www.geeksforgeeks.org/numpy-ones-python/

all_ones = np.ones(len(train_inp), dtype = int)                             # Add all ones for all the hospitalised/inpatient beneficiaries
train_inp['IsHospitalized'] = list(all_ones)                                # Add new feature whether a beneficiary is inpatient or outpatient type

all_zeros = np.zeros(len(train_out), dtype = int)                           # Add all zeros for all the outpatient beneficiaries
train_out['IsHospitalized'] = list(all_zeros)                               # Add new feature whether a beneficiary is inpatient or outpatient type
```

#### <b><u>Observations</b></u>
<b>Why IsHospitalized before merging the dataset?</b>
<ol type = 'a'>
<li>To segregate the Inpatient and Outpatient data.</li>
<li>For patients who were hospitalised, the value is set to 1. This is true for the entire Train InPatient Dataset.</li>
<li>For patients who were not hospitalised, the value is set to 0. This is true for the entire Train OutPatient Dataset.</li>
</ol>

### 3.3 Merge Dataset - Train InPatient Data with Train OutPatient Data

1.png

```python
# Code Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.merge.html

common_lst = []

for col in train_out.columns:                                               # Find common columns in Inpatient and Outpatient Data
  if col in train_inp.columns:
    common_lst.append(col)                                                  # Add the common columns in new list

print(common_lst)

# Merge the Inpatient and Outpatient Data on the common columns list
in_out_patient = pd.merge(train_inp, train_out, left_on = common_lst, right_on = common_lst, how = 'outer')
in_out_patient.head()
```

#### <b><u>Observations</b></u>:
1. Train InPatient Data contains <b>40474 rows</b> and <b>31 columns</b> respectively.
2. Train OutPatient Data contains <b>517737 rows</b> and <b>28 columns</b> respectively.
3. After <u>merging Train Inpatient and Outpatient data</u>, there are <b>558211 rows</b> and <b>31 columns</b> which indicates that <b>28 columns</b> are present in both the datasets.

<b><u>Point of Interest</u>:</b> Unique values of Claim ID Column = Total number of rows in the data. This implies there are <u>no duplicate Claim Id's</u>.

### 3.4 Merge Dataset - Train InPatient/OutPatient Data Merged with Train Beneficiary Data

### 3.8 Plot Individual Percentages of Class Labels

```python
# Count the Class Labels in order to plot them
df1 = train_final.groupby(['PotentialFraud'])['PotentialFraud'].count()                            # Group the Class Labels and then Count them
df1 = df1.reset_index(name='Count')                                                                # Store the Individual Class Label counts into dataframe

# Code Reference: https://www.geeksforgeeks.org/display-percentage-above-bar-chart-in-matplotlib/

percentage = []
total = df1['Count'].sum()
for i in range(df1.shape[0]):
    pct = (df1.Count[i] / total) * 100                                                             # compute percentage of each format
    percentage.append(round(pct,2))
df1['Percentage'] = percentage                                                                     # store the computed percentage into dataframe
print(df1.head(), '\n')

plt.figure(figsize=(5,5))
colors_list = ['Red', 'Green']
graph = plt.bar(df1['PotentialFraud'], df1['Count'], color=colors_list, width=0.3, align='center') # Plot a Bar graph for count of all class labels
plt.title('Count Plot for Fraud and Non Fraud Claims')                                             # Add title of the plot
plt.xlabel("Fraud Status")                                                                         # Add label of x-axis
plt.ylabel("Count")                                                                                # Add label of y-axis
 
i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()                                                                              # Show percentage as text on top and centre of bar graph
    plt.text(x+width/2, y+height*1.01, str(df1.Percentage[i])+'%', ha='center', weight='bold')      
    i += 1
plt.grid()
plt.show()
```

#### <b><u>Observations</b></u>
1. There are <b>212796 Fraudulent claims</b> and <b>345415 Non-Fraudulent claims</b> respectively.
2. Percentage of <b>Fraudulent claims = 38.12%</b> and that of <b>Non-Fraudulent Providers = 61.88%</b> respectively as shown in the above plot.

<b><u>Point of Interest</u>:</b> There is a slight imbalance between the two class labels as per the claims are concerned. This imbalance is <u>less than what we saw for providers and class labels</u>. From the perspective of the analysed data, we can conclude that the dataset is <b>Highly Imbalanced for providers and slightly imbalanced for claims</b>.

## 4. Analysis after addition of new features

### 4.1. Add New Feature - Claim Start Year

```python
# Code Reference: https://www.geeksforgeeks.org/get-month-and-year-from-date-in-pandas-python/

train_final['claim_strt_year'] = pd.DatetimeIndex(train_final['ClaimStartDt']).year    # Add new feature claim start year of the admitted beneficiary
train_final.head()
```

```python
df = pd.DataFrame(train_final.groupby(['claim_strt_year','PotentialFraud'])['BeneID'].count())   # Group the feature for Potential Fraud status
df = df.rename(columns={"BeneID":"Count"})                                                       # Rename the unique feature to Count
df = df.reset_index()                                                                            # Reset Index to get two columns  
total = df['Count'].sum()

percentage = []
for i in range(df.shape[0]):
  pct = (df.Count[i] / total) * 100                                                              # compute percentage for each category
  percentage.append(round(pct,2))
df['Percentage'] = percentage                                                                    # store the computed percentage into dataframe

print(df.head(4), '\n')
plt.figure(figsize= (8, 6))
sns.set_style("ticks",{'axes.grid' : True})                                                      
graph = sns.barplot(x = df['PotentialFraud'], y = df['Count'], hue = df['claim_strt_year'])      # Plot a Bar graph for feature wrt Potential Fraud status
plt.title('Count Plot for Claim Start Year with Potential Fraud Status')                         # Add title of the plot
plt.xlabel('Potential Fraud')                                                                    # Add label of x-axis
plt.ylabel("Count")                                                                              # Add label of y-axis

i = 0
for p in graph.patches:
  width = p.get_width()
  height = p.get_height()
  x, y = p.get_xy()                                                                              # Show percentage as text on top and centre of bar graph
  plt.text(x+width/2, y+height*1.01, str(df.Percentage[i])+'%', ha='center', weight='bold')  
  i += 1
plt.show()
```

#### <b><u>Observations</b></u>
1. <b>37.92%</b> of the <b>98.31% claims</b> that started in <b>2009</b> are <b>fraudulent</b>.
2. Only <b>0.2%</b> of the <b>1.69% claims</b> that started in <b>2008</b> are <b> fraudulent</b>.
3. <b>Why Claim Start Year?</b>
<ol type = 'a'>
<li>To check if any false claim is applied in a particular year.</li> 
<li>It may be the case that all the false claims may have same start year.</li> 
</ol>

<b><u>Point of Interest</u>:</b> From the above plot, it can be seen that the graph for 2008-2009 and potential fraud as no is roughly similar to the graph for 2008-2009 and potential fraud as yes. <u>This implies that claim start year alone cannot be useful in determining if a claim filed in a particular year is potentially fraud or not</u>.

### 5.1 Existing Feature - Diagnostic Codes(Merged)

## 7. Feature Engineering for Feature Interactions

```python
def feature_interaction(train_data, group_col, opr_col, task):
    '''Function takes train data groups it using group_col feature and performs task(eg: mean, count) on opr_col'''
    for val in opr_col:
        new_feature = task+'_'+val+'_'+'per'+''.join(group_col)                         # column name of new feature
        train_data[new_feature] = train_data.groupby(group_col)[val].transform(task)    # group by columns to get the new feature
    return train_data
```
