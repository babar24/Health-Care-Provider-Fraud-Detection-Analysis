## 4. Analysis after addition of new features

```python
# Code Reference: https://www.geeksforgeeks.org/countplot-using-seaborn-in-python/

def bar_plot(feat, head, label1, fig_1):
  '''Plot bar graphs for a feature passed as argument'''
  df1 = train_ben.groupby([feat])[feat].count()                                     # Group the feature and then Count them
  df1 = df1.reset_index(name='Count')                                               # Reset Index to get two columns as Feature name and Count

  # Code Reference: https://www.geeksforgeeks.org/display-percentage-above-bar-chart-in-matplotlib/
  percentage = []
  total = df1['Count'].sum()
  for i in range(df1.shape[0]):
    pct = (df1.Count[i] / total) * 100                                              # compute percentage of each format
    percentage.append(round(pct,2))
  df1['Percentage'] = percentage                                                    # store the computed percentage into dataframe

  plt.figure(figsize= fig_1)
  sns.set_style("ticks",{'axes.grid' : True})
  graph = sns.barplot(x = df1[feat], y = df1['Count'])                              # Plot a Bar graph for a features with respect to its count
  plt.title('Count Plot for '+ head)                                                # Add title of the plot
  plt.xlabel(label1)                                                                # Add label of x-axis
  plt.ylabel("Count")                                                               # Add label of y-axis

  i = 0
  for p in graph.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()                                                               # Show percentage as text on top and centre of bar graph
    plt.text(x+width/2, y+height*1.01, str(df1.Percentage[i])+'%', ha='center', weight='bold')      
    i += 1
  plt.show()
```

```python
# Code Reference: https://seaborn.pydata.org/generated/seaborn.boxplot.html

def box_plot(feat1, head):
  '''Draw box plot for a feature passed as argument'''
  with plt.style.context('seaborn-whitegrid'):                                      # Set the white background for graph plotting
    sns.boxplot(y= feat1, data=train_ben)                                           # Plot Box Plot for the given feature
    plt.grid(linestyle="-", color='grey')                                           # Add grid to the plot
    plt.title('Box Plot for '+ head)                                                # Add title of the plot

# Code Reference: https://www.geeksforgeeks.org/numpy-percentile-in-python/

def percentiles_calculation(feat1):
  '''Claculates 0-100, 90-100 and 99.1-100 percentiles for a feature passed as argument'''
  for i in range(0, 110, 10):                                                  # Calculate 0-100 percentile at an interval of 10 for a given feature
    print(str(i) + ' th percentile is ' + str(np.percentile(train_ben[feat1], i)))
  print('-'*50)
  for i in range(90, 101):                                                     # Calculate 90-100 percentile at an interval of 1 for a given feature
    print(str(i) + ' th percentile is ' + str(np.percentile(train_ben[feat1], i)))
  print('-'*50)
  for i in [99.1,99.2,99.3,99.4,99.5,99.6,99.7,99.8,99.9,100]:                 # Calculate 99-100 percentile at an interval of 0.1 for a given feature
    print(str(i) + ' th percentile is ' + str(np.percentile(train_ben[feat1], i)))
```

### 4.4. Add New Feature - Age of Beneficiary from DOB and DOD

```python
# Code Reference: https://www.geeksforgeeks.org/convert-birth-date-to-age-in-pandas/

out = pd.DatetimeIndex(train_ben.DOD).max()                                 # Calculate the max date from DOD column

train_ben['Age'] = ((out - pd.DatetimeIndex(train_ben['DOB'])).days)/365    # Calculate age from max DOD and DOB column
train_ben['Age'] = train_ben['Age'].round().astype(int)                     # convert age into int
train_ben.head()
```

```python
# Code Reference: https://seaborn.pydata.org/generated/seaborn.histplot.html

with plt.style.context('seaborn-white'):                                                  # Set the white background for graph plotting
  plt.figure(figsize=(20, 8))
  fig = sns.histplot(train_ben['Age'], kde = False, color ='blue', bins = 25)             # Plot Histplot for age feature
  for p in fig.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()                                                                     # Show percentage as text on top and centre of bar graph
    fig.annotate(f'{str(round((height*100)/train_ben.shape[0],2))+"%"}', (x + width/2, y + height*1.01), ha='center', weight='bold')
  plt.grid(linestyle="-", color='grey')                                                   # Add grid to the plot
  plt.title("Count Plot for Age of Beneficiaries")                                        # Add title of the plot
```

#### <b><u>Observations</b></u>
1. <b>Most</b> of the beneficiaries have <b>age</b> between <b>65-93</b>.
2. Least number of beneficiaries have <b>age</b> between <b>26-46</b>.
3. <b>Why Age?</b>
<ol type = 'a'>
<li>To check if any age is written to avail claims.</li> 
<li>To get more idea about false age, we need to check its impact after merging data with all the other csv files.</li>
<li>For example: Very few people(5-6%) from early 30's can show the sign of Alzheimers but it is very common among people of more than 60 years of age.</li>
<li>So, it can help us detect, people of what age have applied for claims related to Alzheimers?</li>
</ol>

#### <b>Q: What is the average inpatient/outpatient annual reimbursement/deductible amount for a beneficiary with Renal Disease?</b> 

```python
def avg_amount_calculation(feat1, feat2):
  '''Calculate the average amount wrt feature1 passed as an argument'''
  count_feat = train_ben.groupby([feat1])['BeneID'].count()                           # Group the feature and then Count them based on BeneID
  avg_feat_amount = train_ben.groupby([feat1])[feat2].sum()/count_feat                # Compute average amount after grouping it based on feature
  return avg_feat_amount                                                              # return average amount of the feature


def amount_plot(feat1, head):
  '''Plot Bar Graph for dataframe passed as an argument'''
  with plt.style.context('seaborn-white'):                                            # Set the white background for graph plotting
    fig = feat1.plot(kind='bar', color=['green','blue','red','orange', 'green','blue','red','orange'], figsize=(8, 6))
    for p in fig.patches:                                                             # Plot a Bar graph for dataframe
      width = p.get_width()
      height = p.get_height()
      x, y = p.get_xy()                                                               # Show percentage as text on top and centre of bar graph
      fig.annotate(f'{int(round(height,0))}', (x + width/2, y + height*1.015), ha='center', weight='bold')
    plt.xlabel(head + " Status")                                                      # Add label of x-axis
    plt.xticks(ticks=[0,1], labels=['No', 'Yes'], rotation = 0)                       # Name the bars of x-axis
    plt.ylabel("Total Annual Average Amount")                                         # Add label of y-axis
    plt.grid(linestyle="-", color='grey')                                             # Add grid for the plot
    plt.title("Total Annual Average Amount for " + head)                              # Add title of the plot
```

1. The <b>Average Inpatient Annual Reimbursement Amount</b> and the <b>Average Inpatient Annual Deductible Amount</b> irrespective of any chronic disease is <b>zero</b> as seen from the percentile calculations.
2. The <b>Average Outpatient Annual Reimbursement Amount</b> and the <b>Average Outpatient Annual Deductible Amount</b> irrespective of any chronic disease is <b>570 and 170</b> respectively as seen from the percentile calculations.

```python
avg_ip_rem1 = avg_amount_calculation('RenalDiseaseIndicator', 'IPAnnualReimbursementAmt')      # function call
avg_ip_ded1 = avg_amount_calculation('RenalDiseaseIndicator', 'IPAnnualDeductibleAmt')         # function call
avg_op_rem1 = avg_amount_calculation('RenalDiseaseIndicator', 'OPAnnualReimbursementAmt')      # function call
avg_op_ded1 = avg_amount_calculation('RenalDiseaseIndicator', 'OPAnnualDeductibleAmt')         # function call

feat_avg_amount1 = pd.concat([avg_ip_rem1, avg_ip_ded1, avg_op_rem1, avg_op_ded1], axis = 1)   # Concatenate all the average values of amount wrt feature1
feat_avg_amount1.rename(columns={0:'Avg_IPReimburseAmt',1:'Avg_IPDeductAmt',2:'Avg_OPReimburseAmt',3:'Avg_OPDeductAmt'},inplace = True)
  
amount_plot(feat_avg_amount1, 'Chronic Renal Disease')                                         # function call
feat_avg_amount1
```

#### <b><u>Observations</b></u>
From the above graph, it can be concluded that the <b>average inpatient/outpatient annual reimbursement/deductible amount</b> is <b>greater</b> for those actually suffering from <u>chronic renal disease</u>.
